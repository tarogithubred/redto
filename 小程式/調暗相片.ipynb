## 讀入套件

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

## 讀入貓貓圖片

```python
img = Image.open("Cat03.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()
```

## 調暗貓貓

```python
img_arr = np.array(img)
darken = int(255*0.3)
darken_arr = np.where((img_arr-0) < darken, 0, img_arr-darken)
pil_image = Image.fromarray(darken_arr)
plt.imshow(pil_image)
plt.axis('off')
plt.show()
```
