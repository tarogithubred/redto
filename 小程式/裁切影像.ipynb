### 裁切影像

```python
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
```

```python
im = Image.open( "Cat03.jpg" )
cut = im.crop( (700, 300, 1500, 1300) )
cut.thumbnail( (400,400) )
cut.save( "croped.jpg" )
```

```python
file_name='croped.jpg';
img=io.imread(file_name)
plt.figure(1)
plt.imshow(img)
plt.axis('off')
```
