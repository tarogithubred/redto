### 把一張照片貼到另一張上面

```python
from PIL import Image
from skimage import io
import matplotlib.pyplot as plt
```

```python
base_im = Image.open("Cat03.jpg")
float_im = Image.open("resized_cat.jpg")
base_im.paste( float_im, (150, 100) )
base_im.save("pasted_cat.jpg")
```

```python
file_name='pasted_cat.jpg';
img=io.imread(file_name)
plt.figure(1)
plt.imshow(img)
plt.axis('off')
```
