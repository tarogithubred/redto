### 模糊照片

```python
from PIL import Image, ImageFilter
from skimage import io
import matplotlib.pyplot as plt
```

```python
im = Image.open("Cat03.jpg")
for i in range(100): im = im.filter(ImageFilter.BLUR)
im.save("blured_cat.jpg")
```

```python
file_name='Cat03.jpg';
img=io.imread(file_name)
plt.figure(1)
plt.imshow(img)
plt.axis('off')
```

```python
file_name='blured_cat.jpg';
img=io.imread(file_name)
plt.figure(1)
plt.imshow(img)
plt.axis('off')
```
