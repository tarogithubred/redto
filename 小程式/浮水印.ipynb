```python
from PIL import Image, ImageDraw, ImageFont
import matplotlib.pyplot as plt
```

## 讀入圖片

```python
img = Image.open("Cat03.jpg")
plt.imshow(img)
plt.axis('off')
plt.show()
```

## 加入浮水印

```python
draw = ImageDraw.Draw(img)
_font = ImageFont.truetype("jf-openhuninn-1.0.ttf",120)
draw.text((30,30), "貓貓好香", font = _font, fill = 'black')
plt.imshow(img)
plt.axis('off')
plt.show()
```
